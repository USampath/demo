```python
"""
10.Please create a new column named total_bedroom_size. If size 10 or less. It should be quoted as small. 
   If the size is 11 or more but 1000 less, it should be medium, otherwise, it should be large.
"""
```

```python
total_bedroom_size = []
for row in df['total_bedrooms']:
        if row < 11.0 :    total_bedroom_size.append('small')
        elif row < 1000.0:   total_bedroom_size.append('medium')
        elif row > 1000.0:  total_bedroom_size.append('large')
        
        
        else:           total_bedroom_size.append('NA')

 
```

```python
df['total_bedroom_size'] = total_bedroom_size
df
```
